### **Validazione di YOLOv10n**

```python
import numpy as np

newDatasetValN = YOLO("/runs/detect/train/weights/best.pt")
metricsN = newDatasetValN.val(data="dataVal.yaml")
print("\n\nmAP50-95:",metricsN.box.map)  # map50-95
print("mAP-50:",metricsN.box.map50)  # map50
print("mAP-75:",metricsN.box.map75)  # map75
print("Average Precision: ", np.mean(metricsN.box.p))  # precision
print("Average Recall: ",np.mean(metricsN.box.r))  # recall
print("Average F1: ",np.mean(metricsN.box.f1))
```

# **Confronto mAP tra i diversi modelli di YOLO**

```python
import matplotlib.pyplot as plt

metriche = ['mAP-50-95', 'mAP-50', 'mAP-75']
modelli = ['Yolov10n', 'Yolov10s', 'Yolov10m']
valori = np.array([
    [metricsN.box.map, metricsS.box.map, metricsM.box.map],        # Valori per mAP-50-95
    [metricsN.box.map50, metricsS.box.map50, metricsM.box.map50],  # Valori per mAP-50
    [metricsN.box.map75, metricsS.box.map75, metricsM.box.map75]   # Valori per mAP-75
])

num_metriche = len(metriche)
num_modelli = len(modelli)
larghezza_colonna = 0.2
spazio = 0.1
posizioni = np.arange(num_metriche) * (num_modelli * larghezza_colonna + spazio)
fig, ax = plt.subplots(figsize=(10, 6))

for i in range(num_modelli):
    ax.bar(posizioni + i * larghezza_colonna, valori[:, i], larghezza_colonna, label=modelli[i])

ax.set_xlabel('Metriche')
ax.set_ylabel('Valore')
ax.set_title('Confronto delle metriche mAP tra modelli')
ax.set_xticks(posizioni + (num_modelli - 1) * larghezza_colonna / 2)
ax.set_xticklabels(metriche)
ax.legend(title='Modelli')

plt.tight_layout()
plt.show()
plt.close()
```

# **Confronto metriche Precision, Recall e F1 tra i diversi modelli**

```python
metriche = ['AVG.Precision', 'Avg.Recall', 'Avg.F1']
modelli = ['Yolov10n', 'Yolov10s', 'Yolov10m']
valori = np.array([
    [np.mean(metricsN.box.p), np.mean(metricsS.box.p), np.mean(metricsM.box.p)],        # Valori per Precision
    [np.mean(metricsN.box.r), np.mean(metricsS.box.r), np.mean(metricsM.box.r)],  # Valori per Recall
    [np.mean(metricsN.box.f1), np.mean(metricsS.box.f1), np.mean(metricsM.box.f1)]   # Valori per F1-score
])

num_metriche = len(metriche)
num_modelli = len(modelli)
larghezza_colonna = 0.2
spazio = 0.1
posizioni = np.arange(num_metriche) * (num_modelli * larghezza_colonna + spazio)

fig, ax = plt.subplots(figsize=(10, 6))

for i in range(num_modelli):
    ax.bar(posizioni + i * larghezza_colonna, valori[:, i], larghezza_colonna, label=modelli[i])

ax.set_xlabel('Metriche')
ax.set_ylabel('Valore')
ax.set_title('Confronto delle metriche di Precision, Recall e F1-score tra modelli')
ax.set_xticks(posizioni + (num_modelli - 1) * larghezza_colonna / 2)
ax.set_xticklabels(metriche)
ax.legend(title='Modelli')

plt.tight_layout()
plt.show()
plt.close()
```

# **Testing "visuale" su dati mai visti prima**

Le seguenti righe di codice permettono di effettuare il testing dei modelli attraverso le immagini di segnali stradali raccolte dal sottoscritto ai fini del progetto, come descritto nella relazione di quest'ultimo.\
**ATTENZIONE**: devono essere presenti le corrispondenti immagini ground-truth all'interno della cartella del dataset (per questo notebook, ciò è assicurato).

```python
import cv2
from PIL import Image
from google.colab.patches import cv2_imshow
from IPython.display import display, HTML, clear_output
import os

# Visualizza il risultato della predizione sulla i-esima immagine della cartella specificata
folder = "./TrafficSigns/images/test"
i=0

modelTest = YOLO("./runs/detect/train/weights/best.pt")     # use YOLOv10n
#modelTest = YOLO("./runs/detect/train2/weights/best.pt")    # use YOLOv10s
#modelTest = YOLO("./runs/detect/train3/weights/best.pt")    # use YOLOv10m


image_files = sorted([f for f in os.listdir(folder) if f.endswith(('.jpg', '.png', '.jpeg'))])

if 0 <= i < len(image_files):
    image_path = os.path.join(folder, image_files[i])

    # se si vuole semplicemente effettuare la prediction, senza salvare i risultati, e senza impostare soglie di IoU/confidence, eseguire la seguente istruzione
    #results = modelTest("datasets/TrafficSigns/images/test")

    # se per ognuno dei file si vuole conservare la nuova immagine con i bounding box stampati in esso, eseguire invece la seguente istruzione
    #results = modelTest("datasets/TrafficSigns/images/test", save=True)

    # se si vogliono impostare delle soglie di IoU e confidenza per scartare risultatii incerti, eseguire invece il seguente codice:
    results = modelTest(image_path, save=False, iou=0.5, conf=0.5)

    noPredImage = cv2.imread(results[0].path, cv2.COLOR_BGR2RGB)
    noPredIMage = cv2.resize(noPredImage , (350,350))
    image = results[0].orig_img
    boxes = []
    for box in results[0].boxes:
      x1, y1, x2, y2 = box.xyxy[0].tolist()
      confidence = box.conf[0].item()
      class_id = int(box.cls[0].item())
      class_name = results[0].names[class_id]
      boxes.append((x1, y1, x2, y2, confidence, class_name))

    boxes = sorted(boxes, key=lambda x: x[4], reverse=True)

    for (x1, y1, x2, y2, confidence, class_name) in boxes:
      label = f'{class_name} - {round(confidence,2)}'
      cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)

      font = cv2.FONT_HERSHEY_DUPLEX
      font_scale = 0.7
      font_thickness = 1
      (text_width, text_height), baseline = cv2.getTextSize(label, font, font_scale, font_thickness)

      text_x = int(x1 + (x2 - x1 - text_width) / 2)
      text_y = int(y1)-10

      cv2.putText(image, label, (text_x, text_y), font, font_scale, (255, 0, 0), font_thickness)

    cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (350,350))

    groundThruthImage = cv2.imread(results[0].path.replace('/images', '/groundThruthAnnotations'), cv2.COLOR_BGR2RGB)
    groundThruthImage = cv2.resize(groundThruthImage , (350,350))


    (h, w) = image.shape[:2]

    text = "Original"
    (text_width, text_height), baseline = cv2.getTextSize(text, font, font_scale, font_thickness)
    text_x = (w - text_width) // 2
    text_y = text_height + 20
    cv2.putText(noPredImage, text, (text_x, text_y), font, font_scale, (255,255,255), font_thickness)

    text = "Ground-thruth"
    (text_width, text_height), baseline = cv2.getTextSize(text, font, font_scale, font_thickness)
    text_x = (w - text_width) // 2
    text_y = text_height + 20
    cv2.putText(groundThruthImage, text, (text_x, text_y), font, font_scale, (255,255,255), font_thickness)

    text = "Predictions"
    (text_width, text_height), baseline = cv2.getTextSize(text, font, font_scale, font_thickness)
    text_x = (w - text_width) // 2
    text_y = text_height + 20
    cv2.putText(image, text, (text_x, text_y), font, font_scale, (255,255,255), font_thickness)

    # mostra in ordine: immagine originale, box/classi ground thruth,box/classi predetti

    concat = cv2.hconcat([noPredIMage, groundThruthImage, image])
    cv2_imshow(concat)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


else:
    print(f"Index {i} out of bound. {len(image_files)} images found in folder: {folder}.\n Please insert a value from 0 to {len(image_files) - 1}.")
```
